```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
titnaic_df=pd.read_csv("titanic(1).csv")
```

```python
sns.histplot(x='Fare',kde=True,bins=10,data=titnaic_df)
sns.scatterplot(x='Fare',data=titnaic_df)
```

```python
#sns.scatterplot(x='Fare',y='Age',data=titnaic_df)
sns.boxplot(x='Sex',y='Age',hue='Survived',data=titnaic_df)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(titnaic_df['Fare'], bins=20, kde='True')
plt.title("distribution of ticket prizes")
plt.xlabel("fare")
plt.ylabel("frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sex', y='Age', hue='Survived', data=titnaic_df)
plt.title('Age Distribution by Gender and Survival')
plt.xlabel('Gender')
plt.ylabel('Age')
#plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```
