```python
q1 = df['FEmarks'].quantile(0.25)
q3 = df['FEmarks'].quantile(0.75)
iqr = q3 - q1    #Interquartile range
lower_boundry = q1 - (1.5*iqr)
upper_boundry = q3 + (1.5*iqr)
print("IQR", iqr, "Lower and upper boundry ",lower_boundry,upper_boundry )
```

```python
sns.boxplot(df['FEmarks'])
plt.show()
```

```python
outliers = df[(df['FEmarks'] < lower_boundry) | (df['FEmarks'] > upper_boundry)]
```

```python
Box plots and scatter plots are both commonly used visualization tools in data analysis, but they serve different purposes and display different types of information. Here's a comparison of the two:

1. **Box Plot**:
   - **Purpose**: Box plots are used to visualize the distribution, central tendency, and spread of a dataset. They are particularly useful for identifying outliers and comparing distributions between different groups or categories.
   - **Components**: A box plot consists of a rectangular box (the "box") with lines extending from it (the "whiskers"). The box represents the interquartile range (IQR), with the median marked inside. The whiskers extend to the minimum and maximum values within a certain range (usually 1.5 times the IQR from the quartiles), and individual data points beyond the whiskers are considered outliers.
   - **Use Cases**: Box plots are commonly used in exploratory data analysis (EDA) to understand the shape and spread of continuous variables and to compare distributions across different groups or categories in a dataset.

2. **Scatter Plot**:
   - **Purpose**: Scatter plots are used to visualize the relationship between two continuous variables. They display individual data points as dots on a two-dimensional plane, with one variable represented on the x-axis and the other on the y-axis. Scatter plots help to identify patterns, trends, correlations, and outliers in the data.
   - **Components**: Each data point in a scatter plot represents a single observation from the dataset, with its position determined by the values of the two variables being plotted. The overall pattern of the points can reveal the nature and strength of the relationship between the variables.
   - **Use Cases**: Scatter plots are commonly used to explore relationships between variables, such as determining whether there is a linear relationship between two continuous variables, identifying clusters or patterns in the data, and detecting outliers or influential points.

In summary, box plots are useful for visualizing the distribution of a single variable or comparing distributions between groups, while scatter plots are useful for visualizing the relationship between two continuous variables and identifying patterns or correlations in the data. The choice between the two depends on the specific questions you want to answer and the type of data you are working with.
```
